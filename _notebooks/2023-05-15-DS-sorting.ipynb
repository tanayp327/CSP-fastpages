```python
# bubble sorts a list of dictionaries, based on the provided key
def bubbleSort(list, key):
    n = len(list) - 1
    
    comparisons = 0  # track number of comparisons
    swaps = 0  # track number of swaps
    
    for i in range(n):
        swapped = False
        
        for j in range(n-i):
            keyN = list[j].get(key)
            keyN1 = list[j+1].get(key)
            
            comparisons += 1  # increment comparisons
            
            if keyN > keyN1:
                swapped = True
                swaps += 1  # increment swaps
                list[j], list[j + 1] = list[j + 1], list[j]
        
        if not swapped:
            break
    
    print(f"Comparisons ({key}): {comparisons}")
    print(f"Swaps ({key}): {swaps}")


# Implementation of selection sort algorithm
def selectionSort(list, key):
    n = len(list)
    
    comparisons = 0  # track number of comparisons
    swaps = 0  # track number of swaps
    
    for i in range(n-1):
        min_idx = i
        
        for j in range(i+1, n):
            comparisons += 1  # increment comparisons
            
            if list[j].get(key) < list[min_idx].get(key):
                min_idx = j
        
        if min_idx != i:
            swaps += 1  # increment swaps
            list[i], list[min_idx] = list[min_idx], list[i]


if __name__ == "__main__":
    # list/dictionary sample
    list_of_people = [
        {"name": "Risa", "age": 18, "city": "New York"},
        {"name": "John", "age": 63, "city": "San Diego"},
        {"name": "Shekar", "age": 18, "city": "San Francisco"},
        {"name": "Ryan", "age": 21, "city": "Los Angeles"}
    ]
    
    # Get the keys from the first dictionary in the list
    key_row = list_of_people[0].keys()
    
    # Print original list
    print("Original List:")
    print(list_of_people)
    
    # Sort using bubble sort
    for key in key_row:
        bubbleSort(list_of_people, key)
        print(f"\nBubble Sort - Sorted by {key}:")
        print(list_of_people)
    
    # Test with your own list
    my_list = [
        {"name": "Emma", "age": 25, "city": "London"},
        {"name": "James", "age": 32, "city": "Paris"},
        {"name": "Sophia", "age": 20, "city": "Berlin"},
        {"name": "Oliver", "age": 28, "city": "Rome"}
    ]
    
    # Print original list
    print("\nOriginal List (My List):")
    print(my_list)
    
    # Sort using bubble sort
    for key in key_row:
        bubbleSort(my_list, key)
        print(f"\nBubble Sort - Sorted by {key}:")
        print(my_list)
    
    # Sort using selection sort
    for key in key_row:
        selectionSort(list_of_people, key)
        print(f"\nSelection Sort - Sorted by {key}:")
        print(list_of_people)
    
    # Test with your own list
    print("\nOriginal List (My List):")
    print(my_list)
    
    # Sort using selection sort
    for key in key_row:
        selectionSort(my_list, key)
        print(f"\nSelection Sort - Sorted by {key}:")
        print(my_list)
```
